```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

x = np.array([106, 119, 130, 160, 191, 229, 266, 306, 353, 502, 650, 709, 739, 780, 833, 855, 874, 928, 934, 971, 1045, 1133, 1182])
y = np.array([810, 800, 785, 687, 570, 522, 561, 588, 569, 334, 92, 24, 12, 7, 16, 30, 53, 173, 176, 330, 565, 737, 769])
dy = 1 / np.max(y)
dx = 1000 / np.max(x)
x = 100 + x * dx
y = 1 - y * dy

tck = interpolate.splrep(x, y, s=0)
xnew = np.linspace(200, 1100, 1000)
ynew = interpolate.splev(xnew, tck, der=0)
fig, ax = plt.subplots()
ax.plot(x, y, 'ro')
ax.plot(xnew, ynew, 'g')
fig.set_figwidth(12)
fig.set_figheight(10)
ax.grid()
print(interpolate.splev(300, tck, der=0))
print(interpolate.splev(400, tck, der=0))
print(interpolate.splev(500, tck, der=0))
print(interpolate.splev(600, tck, der=0))
print(interpolate.splev(750, tck, der=0))
print(interpolate.splev(900, tck, der=0))
print(interpolate.splev(1000, tck, der=0))
print(interpolate.splev(1100, tck, der=0))
print(tck)


```
